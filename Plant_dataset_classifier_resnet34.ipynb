#### Recreate data bunch based on the clean files

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, bs=32).normalize(imagenet_stats)
```

```python
print('Total number of classes = ', len(data.classes))
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

```python
print('Number of training examples = ', len(data.train_ds))
print('Number of validation samples = ', len(data.valid_ds))
```

### Train model

```python
learn = cnn_learner(data, models.resnet34, metrics = error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1-34')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.load('stage-1-34')
```

```python
learn.fit_one_cycle(10, max_lr= slice(1e-6, 1e-2))
```

```python
learn.save('stage-2-34')
```

```python
learn.fit_one_cycle(10, max_lr= slice(1e-6, 1e-2))
```

### Classifier interpretation

```python
learn.load('stage-2-34')
```

```python
interpret = ClassificationInterpretation.from_learner(learn)
```

```python
interpret.plot_confusion_matrix(figsize = (20,20), dpi=150)
```

```python
interpret.plot_top_losses(9, figsize=(15,11))
```
